## What is Debugging?
Debugging is the process of identifying and fixing errors (bugs) in your code.
Think of it as detective work where you:
1. **Find the problem**: Figure out what's causing your program to behave unexpectedly.
2. **Understand the issue**: Determine why the bug is happening.
3. **Fix the error**: Correct the code to make your program work as expected.



image.png

## Types of Python Errors
Errors are mistakes in your code that prevent it from running correctly. Here are common types of Python errors:
1. **SyntaxError**: Occurs when your code has invalid syntax (e.g., missing a colon).
   - Example: `print("Hello world"`
2. **NameError**: Happens when you use a variable or function that hasn't been defined.
   - Example: `print(variable_that_does_not_exist)`
3. **TypeError**: Occurs when you perform an operation on incompatible data types.
   - Example: `3 + "hello"`
4. **IndexError**: Happens when you try to access an index that doesn‚Äôt exist in a list.
   - Example: `my_list = [1, 2, 3]; print(my_list[5])`

## Functions

Functions are an essential part of Python and programming in general. They allow you to group a set of instructions into a reusable block of code. By using functions, you can break down a large program into smaller, more manageable pieces, making it easier to read, write, and debug your code.

It's a block of code that:

* Has a specific name
* Can take information (called "inputs" or "parameters")
* Does something with that information
*Usually gives back a result (called a "return value")

image.png

### Creating a Function

To create a function in Python, you use the `def` keyword. Here's the basic structure of a function:


````python
def function_name():
    # Code that the function will execute
````

üîç Let's break this down:

- `def` is the keyword used to define a function.
- `function_name` is the name you give your function. This should describe what the function does.
- The colon (`:`) indicates the start of the function's code block.
- The code inside the function is indented (usually 4 spaces) to show it belongs to the function.

Here‚Äôs a basic function that prints a greeting message:

````python
def greet():
    print("Hello, world!")
````

To **call** the function (or make it run), you use its name followed by parentheses:

````python
# Call the function
greet()
````

**Understanding Function Names**

- Function names should be descriptive so that others can understand their purpose.
- Use lowercase letters and underscores (`_`) to separate words, like `calculate_sum` or `greet_user`.
- Avoid using Python‚Äôs reserved keywords (like `print`, `if`, `else`, etc.) as function names.

### Functions with Parameters

Functions can take inputs, called **parameters**, to make them more versatile. For example, a function can display a customized message:


````python
def greet_person(name):
    print("Hello,", name)
````

````python
# Call the function with different names
greet_person("Alex")
greet_person("Taylor")
````

üîç Let's break this down:
- `def` tells Python you're creating a function
- `greet` is the name we gave our function
- `(name)` is the input our function needs
- Everything indented after the `:` is what the function does

### Returning Values from Functions

Sometimes, you want a function to calculate or process something and then give back a result. You can do this with the `return` statement.

````python
def add_numbers(a, b):
    return a + b
````

````python
# Store the result in a variable
result = add_numbers(5, 3)
print(result)  # Outputs: 8
````

### Functions with Multiple Parameters

A function can have more than one parameter. For example:

````python
def calculate_area(length, width):
    return length * width
````

You can call this function like so:

````python
area = calculate_area(5, 3)
print("The area is:", area)
````

### Using Docstrings in Python

A **docstring** (short for documentation string) is a special kind of string in Python that describes what a function, class, or module does. It is a great way to explain the purpose, inputs, outputs, and behavior of your code, making it easier for others (and yourself) to understand your work later.

#### **How to Write a Docstring**
- Docstrings are enclosed in triple quotes (`"""` or `'''`).
- They are written as the first statement in a function, class, or module.
- A well-written docstring should include:
  1. **A brief description of the function‚Äôs purpose.**
  2. **Information about parameters (if any).**
  3. **Details about the return value (if any).**

#### **Example of a Function with a Docstring**
Here‚Äôs a function that calculates the area of a rectangle, with a proper docstring:

````python
def calculate_area(length, width):
    """
    Calculate the area of a rectangle.

    Parameters:
    length (float): The length of the rectangle.
    width (float): The width of the rectangle.

    Returns:
    float: The area of the rectangle.
    """
    return length * width
````

**How to View a Docstring**

You can view a function's docstring by using the help() function:

````python
help(calculate_area)
````

Using docstrings is a professional practice that ensures your code is well-documented and easy to maintain!

## **Dictionaries**

image.png


A **dictionary** in Python is a collection of key-value pairs. It is used to store data where each piece of information (value) is labeled with a unique identifier (key). Think of a dictionary as a real-world dictionary, where you look up words (keys) to find their meanings (values).

#### **Key Features of Dictionaries**:
- **Mutable**: You can change, add, or remove items.
- **Keys are unique**: No duplicate keys are allowed.
- **Fast Lookups**: They provide fast access to data when you know the key.



### **Creating a Dictionary**

You create a dictionary using curly braces `{}` and separate keys and values with a colon (`:`). Multiple key-value pairs are separated by commas.

```python
# Example: Dictionary of a student‚Äôs details
student = {
    "name": "Alice",
    "age": 20,
    "course": "Computer Science",
    "grade": "A"
}
print(student)



### Why do we need dictionaries?
Here‚Äôs why dictionaries are super useful:

1. **Quick Lookup**:
   - Imagine you have a list of student names and their grades. If you use a list, you‚Äôd have to scan through the entire list to find someone‚Äôs grade. With a dictionary, you can directly use the student‚Äôs name (the key) to get their grade (the value) instantly.

   **Example**:
   ```python
   grades = {"Alice": 95, "Bob": 87, "Charlie": 78}
   print(grades["Alice"])  # Output: 95
   ```

2. **Organized Data**:
   - Dictionaries help organize your data in a way that makes sense. Instead of keeping separate lists for names, ages, and addresses, you can group everything together in a dictionary for each person.

   **Example**:
   ```python
   person = {
       "name": "Alice",
       "age": 16,
       "address": "123 Maple Street"
   }
   print(person["address"])  # Output: 123 Maple Street
   ```

3. **Flexibility**:
   - Dictionaries let you store all kinds of information‚Äînumbers, strings, lists, or even other dictionaries! This makes them powerful for managing complex data.

4. **No Fixed Order**:
   - Unlike lists where items are stored in a specific order, dictionaries don‚Äôt rely on order. You can access any key-value pair directly without worrying about where it‚Äôs stored.


### Why not just use lists?
Let‚Äôs compare:

| **Lists**                           | **Dictionaries**                     |
|-------------------------------------|--------------------------------------|
| Store items in a sequence.           | Store data in key-value pairs.       |
| Access by position (index).          | Access by key.                       |
| Harder to look up specific values.   | Easy to look up values by their keys.|

**Understanding Keys and Values**

**Keys**:
- The keys in a dictionary must be **unique**.
- Keys can be strings, numbers, or other immutable data types.


````python
my_dict = {
    "name": "Alice",  # Key is a string
    101: "Student ID",  # Key is a number
}

print(my_dict)
````

**Values**:
- Values can be **any type of data**: strings, numbers, lists, or even other dictionaries.

````python
my_dict = {
    "age": 20,  # Value is a number
    "hobbies": ["reading", "cycling"],  # Value is a list
}

print(my_dict)
````

### **Creating an Empty Dictionary**

You can also start with an empty dictionary and add items to it later:

````python
# Creating an empty dictionary
student = {}

# Adding key-value pairs
student["name"] = "Bob"
student["age"] = 22
student["grade"] = "B"

print(student)
````

### **Accessing Items in a Dictionary**

To retrieve a value from a dictionary, use its key inside square brackets (`[]`):

````python
print(student["name"])  # Output: Bob
print(student["age"])   # Output: 22
````

### **Adding and Updating Items in a Dictionary**

You can add new key-value pairs or update existing ones:

````python
# Adding a new key-value pair
student["hobby"] = "Reading"
print(student)
````

````python
# Updating an existing key
student["grade"] = "A"
print(student)
````

**Understanding keys(), values(), and items() in Dictionaries**

Dictionaries in Python store data as key-value pairs. To work with these pairs, Python provides the methods .keys(), .values(), and .items(). Here's how they work:

**1. keys()**

The .keys() method returns a view of all the keys in the dictionary.

Example:

````python
fruits = {"apple": 3, "banana": 5, "cherry": 2}

print(fruits.keys())

# Looping through the keys:
for key in fruits.keys():
    print(key)
````

**2. values()**

The .values() method returns a view of all the values in the dictionary.

````python
fruits = {"apple": 3, "banana": 5, "cherry": 2}

print(fruits.values())

# Looping through the values:
for value in fruits.values():
    print(value)
````

**3. items()**

The .items() method returns a view of all key-value pairs.

````python
fruits = {"apple": 3, "banana": 5, "cherry": 2}

print(fruits.items())

# Looping through key-value pairs:
for key, value in fruits.items():
    print(key,value)
````
